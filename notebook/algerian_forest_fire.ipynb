### **Algerian Forest Fires Dataset**

Data Set information

The dataset include 244 instances that regroup a data of two regions of Algeria, namely the Bejaia region located in the northest of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region

The period from june 2012 to September 2012. The dataset includes 11 attribues and 1 output attribue (class). The 244 instances have been classified into fire (138 classes) and fire (106 classes) classes.

Attribute information:
1. Date : (DD/MM/YYYY) Dat, month(june to september), year (2012) Weather data observations
2. Temp : temperature noon (temperature max) in Celsius degrees : 22 to 42
3. RH : Relvative Humidity in %: 21 to 90
4. Ws : Wind speed in km/h : 6 to 29
5. Rain : total day in mm : 0 to 16.8 FWI Components
6. Fire Fuel Moisture Code (FFMC) : index from the FWI system : 28.6 to 92.5
7. Duff Moisture Code(DMC) : index from the FWI system : 1.1 to 65.9
8. Drought Code (DC): index from the FWI system:7 to 220.4
9. initial Spread Index (ISI) : index from the FWI system : 0 to 18.5
10. Buildup Index (BUI) : index from the FWI system: 1.1 to 68
11. Fire Weather index (FWI) index : 0 to 31.1
12. Classes: two classes namely Fire and not Fire

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', header=1)
dataset.head()
```

```python
dataset.info()
```

**Data Cleaning**

```python
# missing value
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on Region from 122th index, we can make a new column based on the Region
1. "Bejaia Region Dataset"
2. "Sidi-Bel Abbes Region Dataset"

Add new column with region

```python
dataset.loc[:122, "Region"] = 0
dataset.loc[122:, "Region"] = 1
df = dataset
```

```python
df.info()
```

```python
df.head()
```

```python
df['Region'] = df['Region'].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# removing the null values
df = df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
# fix spaces in columns names
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
df.head()
```

```python
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature']] = df[[
    'day', 'month', 'year', 'Temperature']].astype(int)
```

```python
df['Ws'] = df['Ws'].astype(int)
```

```python
df[['RH']] = df[['RH']].astype(int)
```

```python
df.info()
```

```python
df.head()
```

**Change the other columns to float data datatype**

```python
object = [feature for feature in df.columns if df[feature].dtype == 'O']
object
```

```python
for i in object:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```
